```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
import numpy as np 
```

```python
df= pd.read_csv('/datasets/users_behavior.csv')
df.info()
print(df.head(10))
```

```python
features = df.drop(['is_ultra'], axis = 1)
target = df["is_ultra"]
features_temp, features_test, target_temp, target_test = train_test_split(features, target, test_size = 0.2, random_state = 54321)
features_train, features_valid,target_train, target_valid = train_test_split(features_temp, target_temp, test_size = 0.25, random_state =54321 )
```

```python
for depth in [5,10, 15] :
    model_1 = DecisionTreeClassifier(random_state= 54321, max_depth = depth)
    model_1.fit(features_train, target_train)
    model_1_prediction = model_1.predict(features_valid)
    accuracy_model_1 = accuracy_score(target_valid,model_1_prediction)
    print(f"depht {depth} ,porcentaje de acierto de : {accuracy_model_1}")
```

de este modelo, el arbol de decision con el depth de 5 tuvo el mayor porcentaje de exactitud con un 81.8% 

```python
for estimators in [10, 50, 100] :
    model_2 = RandomForestClassifier(random_state = 54321, n_estimators = estimators)
    model_2.fit(features_train, target_train)
    model_2_prediction =model_2.predict(features_valid)
    accuracy_model_2 = accuracy_score(target_valid, model_2_prediction)
    print(f"n_estimators ={estimators}, accuracy ={accuracy_model_2}")
```

```python
model_3 = LogisticRegression()
model_3.fit(features_train, target_train)
model_3_prediction = model_3.predict(features_valid)
accuracy_model_3 = accuracy_score(target_valid, model_3_prediction)
print('Accuracy = ', accuracy_model_3)
```

Tal como se puede observar el modelo con mayor exactitud es el modelo de RandomForestClassifier cuando n estimators = 50 dando un 83% de exactitud

```python
best_model = RandomForestClassifier(random_state = 54321, n_estimators = 50)
best_model.fit(features_train, target_train)
best_model_prediction = best_model.predict(features_test)
accuracy_best_model= accuracy_score (target_test , best_model_prediction)
print("la exactitud del modelo es de ",accuracy_best_model  )
```

```python
print('Clases unicas', target_test.unique() )
print('Cantidad de clases :', len(target_test.unique()))
random_predictions = np.random.choice(target_test.unique(), size = len(target_test))
accuracy_random = accuracy_score(target_test, random_predictions)
print("Si hicieramos aleatoriamente la prediccion, su exactitud seria del ",accuracy_random)
```

haciendo la puntuacion de la exactitud con los datos de la prueba, se confirmo que cumple con el umbral antes establecido de mas de 75% de exactitud con un total de 78.7% de exactitud, superando la prueba de cordura por 30 puntos porcentuales  , demostrando que este modelo es el de mayor exactitud
